```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from fairlearn.datasets import fetch_adult
```

```python
data = fetch_adult(as_frame=True)
df = data.frame.copy()

df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())

df.info()
```

```python
TARGET = "income"
PROTECTED = ["sex", "race"]

print("Target:", TARGET)
print("Protected Attributes:", PROTECTED)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df[TARGET].value_counts(normalize=True)

sns.countplot(x=TARGET, data=df)
plt.title("Target Distribution (Income)")
plt.show()
```

```python
df["sex"].value_counts(normalize=True)

sns.countplot(x="sex", data=df)
plt.title("Sex Distribution")
plt.show()
```

```python
df["race"].value_counts(normalize=True)

plt.figure(figsize=(8,4))
sns.countplot(x="race", data=df)
plt.xticks(rotation=30)
plt.title("Race Distribution")
plt.show()
```

```python
df["intersectional"] = df["sex"] + "_" + df["race"]

df["intersectional"].value_counts(normalize=True)
```

```python
pd.crosstab(df["sex"], df[TARGET], normalize="index")
pd.crosstab(df["race"], df[TARGET], normalize="index")
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
